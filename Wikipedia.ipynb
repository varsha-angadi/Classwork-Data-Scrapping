```python
# Import required libraries

import requests

from bs4 import BeautifulSoup

import pandas as pd
```

```python
# Set up headers to mimic a browser visit

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'
}
```

```python
# URL of the Wikipedia page of 'List of states and territories of the United States' 

url='https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
```

```python
#sending GET request

response=requests.get(url)
```

```python
# Parse the HTML content

soup = BeautifulSoup(response.content,'html.parser')
```

```python
# Raw HTML content

soup
```

```python
# Locate and extract the first table element with the class attribute 'wikitable'

table = soup.find('table', class_='wikitable')
```

```python
table
```

```python
# Find and return a list of all <th> elements (table headers) within the 'table'

table.find_all('th')
```

```python
# Find and return a list of all <tr> elements (table rows) within the 'table'

table.find_all('tr')
```

```python
# Find and return a list of all <tr> elements (table headers) within the 'table' starting from second index to skip headers

for i in table.find_all('tr')[2:]:
    print(i)
    print("----")
```

```python
# # Initialize an empty list to store state names
States=[]
# Iterate over all rows in the table, starting from the third row (index 2) to skip headers
for i in table.find_all('tr')[2:]:
    # Find the first <a> tag within the row and extract its text, then append it to the States list
    States.append(i.find('a').text)
```

```python
States
```

```python
Shortname=[]

for i in table.find_all('tr')[2:]:
    Shortname.append(i.find('td').text)
```

```python
Shortname
```

```python
Population=[]

for i in table.find_all('tr')[2:]:
    Population.append(i.find('div').text)
```

```python
Population
```

```python
# Create a DataFrame

df=pd.DataFrame({'States':States, 'Abbreviation':Shortname, 'Population':Population})
```

```python
df
```

```python
# Save to CSV

df.to_csv('State-wise_Poluation_USA.csv', index=False)
```
